```python
import matplotlib.pyplot as plt

university_ranking = {"Tribhuvan Uni": 23,
                      "Kathmandu Uni": 48,
                      "Pokhara Uni": 36}

fig, ax = plt.subplots()
bars = ax.bar(university_ranking.keys(), university_ranking.values())

# Invert the y-axis
ax.invert_yaxis()

# Add labels to bars
ax.set_ylabel('Ranking (Lower is Better)')
ax.set_xlabel('Universities')
ax.set_title('University Rankings')

plt.show()
```
